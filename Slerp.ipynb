```python
import yaml

MODEL_NAME = "Marcoro14-7B-slerp"
yaml_config = """
slices:
  - sources:
      - model: Bilal326/gpt2-finetuned-imdb-sentiment
        layer_range: [0, 32]
      - model: Bilal326/DistilBERT-finetuned-imdb-sentiment
        layer_range: [0, 32]
merge_method: slerp
base_model: Bilal326/DistilBERT-finetuned-imdb-sentiment
parameters:
  t:
    - filter: self_attn
      value: [0, 0.5, 0.3, 0.7, 1]
    - filter: mlp
      value: [1, 0.5, 0.7, 0.3, 0]
    - value: 0.5
dtype: bfloat16

"""

# Save config as yaml file
with open('config.yaml', 'w', encoding="utf-8") as f:
    f.write(yaml_config)
```
